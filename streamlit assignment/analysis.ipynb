```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import streamlit as st
```

```python
df= pd.read_csv("vgsales.csv")
df
```

```python
df.isnull().sum()
```

```python
df1=df.dropna()
df1
```

```python
df1.info()
```

```python
df1.to_csv("Vgsales1.csv",index=False)
```

```python
df1
```

```python
df1.groupby("Genre")["Global_Sales"].sum().reset_index().sort_values(by="Global_Sales",ascending=False).head()
px.bar(df1,x="Genre",y="Global_Sales",color="Genre")
```

```python
df1.groupby("Genre")["NA_Sales"].sum().reset_index().sort_values(by="NA_Sales",ascending=False).head()
px.bar(df1,x="Genre",y="NA_Sales",color="Genre")
```

```python
df1.groupby("Genre")["EU_Sales"].sum().reset_index().sort_values(by="EU_Sales",ascending=False).head()
px.bar(df1,x="Genre",y="EU_Sales",color="Genre")
```

```python
df1.groupby("Genre")["JP_Sales"].sum().reset_index().sort_values(by="JP_Sales",ascending=False).head()
px.bar(df1,x="Genre",y="JP_Sales",color="Genre")
```

```python
df1.groupby("Genre")["Other_Sales"].sum().reset_index().sort_values(by="Other_Sales",ascending=False).head()
px.bar(df1,x="Genre",y="Other_Sales",color="Genre")
```
