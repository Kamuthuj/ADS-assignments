```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("jumta.csv")
df
```

```python
# Create a line plot showing the monthly revenue over the two-year period. Customize the plot to make it publication-quality.
```

```python
plt.plot(df["Month"],df["Revenue"],marker=".",linestyle="-",linewidth=1,label="Revenue")
plt.xlabel("month")
plt.ylabel("revenue")
plt.xticks(rotation=45)
plt.title("Monthly revenue btwn Jan 2020 to Dec 2021")
plt.grid()
plt.legend()
plt.show()
plt.savefig("Monthly revenue.png")
```

```python
#  Create a bar plot showing the total number of orders per month. Customize the plot to make it publication-quality
```

```python
df
```

```python
plt.bar(df["Month"],df["Orders"], color="black",label="orders")
plt.title("Monthly orders")
plt.xlabel("Month")
plt.ylabel("Orders")
plt.xticks(rotation=45)
plt.legend()
plt.show()
plt.savefig("monthly orders barplot.png")
```

```python
# Create a scatter plot showing the relationship between the number of orders and the average order value. 
# Customize the plot to make it publication-quality.
```

```python
df
```

```python
plt.scatter(df["Orders"],df["Average_Order_Value"],color="black",label ="av.order value")
plt.xlabel = "orders"
plt.ylabel ="order value"
plt.title ="Order value"
plt.grid()
plt.legend()
plt.show()
plt.savefig("order value scatterplot.png")
```
