```python
data1=df.dropna()
data1
```

```python
data1.isnull().sum()
```

```python
# 5 Display the top 5 video games by global sale
```

```python
data1.groupby("Name")["Global_Sales"].sum().reset_index().sort_values(by="Global_Sales",ascending=False).head(5)
```

```python
# 6 Calculate the total sales in North America for the 'Action' genre
```

```python
data1
```

```python
data1[data1["Genre"]=="Action"]["NA_Sales"].sum()
```

```python
# 7 Find the top 3 publishers by global sales.
```

```python
data1.groupby("Publisher")["Global_Sales"].mean().reset_index().sort_values(by="Publisher",ascending=False).head(3)
```

```python
# 8 Calculate the average sales in Europe for the 'Shooter' genre.
```

```python
data1[data1["Genre"]=="Shooter"]["EU_Sales"].mean()
```

```python
#  9 Find the game with the highest sales in Japan in the 'Sports' genre.
```

```python
data1.groupby("Genre")["JP_Sales"].sum().reset_index().sort_values(by="Genre",ascending=False).head(1)
```
